```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error
from chefboost.training import Training
from chefboost import Chefboost as chef
```

```python
filename='Bank_Note.csv'
names = ['variance', 'skewness', 'curtosis', 'entropy','Decision']
XX = ['variance', 'skewness', 'curtosis', 'entropy']
df=pd.read_csv(filename,names=names)
df
```

```python
ds=df.replace(r'^\s*$', np.nan, regex=True)
ds
```

```python
for i in XX:
    ds[i]=pd.to_numeric(ds[i],downcast="float")
ds
```

```python
ds.isnull().sum()
```

```python
ds=ds.fillna(ds.mean())
```

```python
ds
```

```python
X=ds.iloc[:,0:4]
Y=ds.iloc[:,4]
Input=X.values
Output=Y.values
#splitting the data into 20% test data and 80% training data.
input_train, input_test, output_train, output_test = train_test_split(Input, Output, test_size = 1/5, random_state = 1)
```

```python
print(input_train)
print("\n")
print(input_train)
```

```python
print(output_train)
print("\n")
print(output_test)
```

```python
model=DecisionTreeClassifier()   # CART
model.fit(input_train,output_train)
predicted_output = model.predict(input_test)
#print(predicted_output)
mm=len(predicted_output)
#print(output_test)
#print(len(output_test))
count=0
for i in range(len(output_test)):
    if output_test[i]==predicted_output[i]:
        count+=1
    else:
        pass
l=(count/mm)*100
print("Accuracy for the Test Data-{0}".format(l))
```
