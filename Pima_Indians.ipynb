```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error
```

```python
filename='Pima_indians_diabetics.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
df=pd.read_csv(filename,names=names)
df
```

```python
ds=df.replace(r'^\s*$', np.nan, regex=True) #removing all the voids and replacing with NaN
```

```python
ds.isnull().sum()
```

```python
d1=ds.fillna(method='ffill') # filling the dataset using forwardfill and filing all the Nan place. 
```

```python
d1        # final dataset
```

```python
ds1=d1.interpolate()
```

```python
ds1
```

```python
ds1.isnull().sum()
```

```python
X=ds1.iloc[:,0:8]
Y=ds1.iloc[:,8]
Input=X.values
Output=Y.values
#splitting the data into 20% test data and 80% training data.
input_train, input_test, output_train, output_test = train_test_split(Input, Output, test_size = 1/5, random_state = 1)
```

```python
print(input_train)
print(output_train)
```

```python
model = GaussianNB()    # model selection - Gaussian Naive Bayes
model.fit(input_train,output_train)    # model Training 
predicted_output = model.predict(input_test)     # model testing using test data
print(predicted_output)
VarianceGB = np.var(predicted_output)       # variance 
test_errorGB = mean_squared_error(output_test,predicted_output)       # squared error which was same as mean Squared error in the previous line
BiasGB=np.sqrt(test_errorGB+VarianceGB)     # MSE=(Bias)^2-Variance
```

```python
#input from the user or any device
preg =float(input("preg->"))
plas =float(input("plas->"))
pres =float(input("pres->"))
skin =float(input("skin->"))
test =float(input("test->"))
mass =float(input("mass->"))
pedi =float(input("pedi->"))
age  =float(input("age->"))
testinput = [[preg,plas,pres,skin,test,mass,pedi,age]]
predicted_output=model.predict(testinput)
print(predicted_output)
```

```python
kfold = KFold(n_splits=10, random_state=7)
resultsGB = cross_val_score(model, X, Y, cv=kfold)
print(resultsGB.mean())  # cross validation score for Gaussian Naive Bayes
```

```python
model = KNeighborsClassifier() # model selection - Knn
model.fit(input_train,output_train)  # model Training
predicted_output=model.predict(input_test) # predicted output from the test input
print(predicted_output)
VarianceKN = np.var(predicted_output)       # variance 
test_errorKN = mean_squared_error(output_test,predicted_output) # squared error which was same as mean Squared error in the previous line
BiasKN=np.sqrt(VarianceKN+test_errorKN)  # MSE=(Bias)^2-Variance
```

```python
#Input from a user or a device
preg =float(input("preg->"))
plas =float(input("plas->"))
pres =float(input("pres->"))
skin =float(input("skin->"))
test =float(input("test->"))
mass =float(input("mass->"))
pedi =float(input("pedi->"))
age  =float(input("age->"))
testinput = [[preg,plas,pres,skin,test,mass,pedi,age]]
predicted_output=model.predict(testinput)   # predicting to a user input 
print(predicted_output)
```

```python
kfold = KFold(n_splits=10, random_state=7)
resultskn = cross_val_score(model,X,Y,cv=kfold)
```

```python
#from Cross Validation we can see the crossVal of Knn< crossVal of GB
```

```python
print("Cross Validation score for Naive Bayes: {0}".format(resultsGB.mean()))# cross validation score for Naive Bayes
print("Variance for Gausian Naive Bayes : {0}".format(VarianceGB))
print("Bias for Gausian Naive Bayes: {0}".format(BiasGB))
#-------------------------------------------------------------------------------------------------------
print("Cross Validation score for KNN: {0}".format(resultskn.mean()))# cross validation score for Knn
print("Variance for KNN : {0}".format(VarianceKN))
print("Bias for KNN: {0}".format(BiasKN))
```
